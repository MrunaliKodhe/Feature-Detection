```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load image in grayscale
img = cv2.imread('/content/architecture-as-a-product-what-makes-a-building-worth-repeating_1.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Harris Corner Detector
gray = np.float32(gray)
harris_corners = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)

# Dilate corner points for better visualization
harris_corners = cv2.dilate(harris_corners, None)

# Mark strong corners in red
img[harris_corners > 0.01 * harris_corners.max()] = [0, 0, 255]

# Display the result
plt.axis("off")
plt.imshow(img)
```
